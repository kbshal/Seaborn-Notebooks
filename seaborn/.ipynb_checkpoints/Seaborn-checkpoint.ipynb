## In this session we are going to learn and practise seaborn

## Some terms

#### 1.  If we have a one item to determine like only f1 or only one axis then it is called <font color="red" size=4px>Univariant</font> analysis 
#### 2. If we have two items or two axis to determine or analyse then it is called <font color="red" size=4px>Bivariant</font> analysis

### Distribution plots
- distplot
- jointplot
- pairplot


```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

### Correlation with Heatmap

A correlation heatmap uses colored cells, typically in monochromatic scale, to show 2D correlation matrix (table) between two discrete event type. It is very important in feature selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

- Joinplot
A joinplot helps you to study the realtion between two numeric values. The central chart displays their correlation. It is usally a scatter plot,a hexbin plot,2D histgram  or 2D density plot

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="hex")
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="reg")
```

#### Pair Plot

pair plot is used to analyse the bivariant features. Like if you two or more than two independent features pairplot will help you to analyse or visualize those features

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

- <font color='red' size=4px> Distplot</font>

Distplot is used to create the histogram 

```python
sns.distplot(df["tip"])
```

```python
sns.distplot(df["tip"],kde=False,bins=10)
```

### Categorical Plots
- box plot
- violin plot
- count plot
- bar plot 

```python
# count plot

sns.countplot('sex',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.barplot(y="total_bill",x="sex",data=df)
```

### Box plot

A box or whisker plot( sometimes called box plot) is a graph that represents the informatiion from five summary numbers

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

### Violin plot

violin plot is a graph which helps us to see both distribution in the kernel density distribution and box plot


```python
sns.violinplot('sex','total_bill',data=df)
```
